```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from sklearn.ensemble import RandomForestRegressor

def CreateDataset(fName, size, seed):
    np.random.seed(seed)
    
    N = size
    
    X1 = np.random.rand(N)
    X2 = np.random.rand(N)
    X3 = np.random.rand(N)
    X_Sum = X1+X2+X3
    X1 = X1/X_Sum
    X2 = X2/X_Sum
    X3 = X3/X_Sum
    
    Y = np.zeros((N,))
    # logic      
    for i in range(N):
        noise = np.random.rand()
        Y[i] = 0.5*X1[i] + 3*X2[i] + 5*X3[i] + 5*X2[i]*X3[i] + 2*X1[i]*X2[i]*X3[i]  + noise
              
  
            
    data = np.array([X1, X2, X3 ,Y]).T
    df = pd.DataFrame(data,columns=['radio','tv','internet','sales'])
    return df

if __name__ == "__main__":
    #Leave the original code unchanged
    df_train = CreateDataset("synthetic_sales_train.csv", 1000, 1)  
    df_test = CreateDataset("synthetic_sales_test.csv", 1000, 2)  
    
    #Split Data
    X_train = df_train[['radio','tv','internet']]
    y_train = df_train[['sales']].values.reshape(-1,) 
    X_test = df_test[['radio','tv','internet']]
    y_test = df_test[['sales']].values.reshape(-1,)
    
    #Ordinary Least Square
    #1. fit the model
    #2. predict
    #3. calculate the MSE error
    lin_reg = sm.OLS(y_train, X_train).fit() 
    print(lin_reg.summary(),end="\n\n\n")
    y_hat = lin_reg.predict(X_test)
    print("***********************************************************")
    print("Regression mean squared error(testing): ", np.mean(np.power((y_test- y_hat).values,2)),end="\n\n")
    
    #Random Forest
    #1. fit the model
    #2. predict
    #3. calculate the MSE error
    reg = RandomForestRegressor(500, random_state=0)
    reg.fit(X_train, y_train)
    y_hat = reg.predict(X_test)
    #print("***********************************************************")
    print("random forest loss(testing): ", np.mean(np.power((y_test- y_hat),2)),end="\n\n")
    
    
    #Compute tow models above's training error
    print("***********************************************************",end="\n\n")
    y_hat = lin_reg.predict(X_train)
    print("regression (training) mean squared error: ", np.mean(np.power((y_train-y_hat).values,2)),end="\n\n")
    y_hat = reg.predict(X_train)
    print("random forest(training) loss: ", np.mean(np.power((y_train- y_hat),2)))
    print("***********************************************************")  
```
